# **Apriori Algorithm**

```python
from mlxtend.frequent_patterns import apriori
import pandas as pd

# Simple shopping data
data = pd.DataFrame({
    'Chips': [1,1,1,0,1],
    'Soda':  [1,0,1,1,1],
    'Candy': [0,1,0,1,0]
})

# Find frequent itemsets
result = apriori(
    data,
    min_support=0.6,
    use_colnames=True
)

print(result)
```

```python
What This Code Does
1 means the item was bought
0 means it was not bought
min_support=0.6 means:
the item must appear in at least 60% of transactions

The program automatically finds popular item combinations.
```

```python
import pandas as pd

transactions = [
    ['Apple', 'Banana', 'Cookie'],
    ['Apple', 'Banana'],
    ['Apple', 'Cookie'],
    ['Banana', 'Cookie'],
    ['Apple', 'Banana', 'Cookie']
]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(
    te_array,
    columns=te.columns_
)

print(df)
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets = fpgrowth(
    df,
    min_support=0.4,
    use_colnames=True
)

print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

print(
    rules[
        ['antecedents',
         'consequents',
         'support',
         'confidence',
         'lift']
    ]
)
```
