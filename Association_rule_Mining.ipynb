# **Apriori Algorithm**

```python
from mlxtend.frequent_patterns import apriori
import pandas as pd

# Simple shopping data
data = pd.DataFrame({
    'Chips': [1,1,1,0,1],
    'Soda':  [1,0,1,1,1],
    'Candy': [0,1,0,1,0]
})

# Find frequent itemsets
result = apriori(
    data,
    min_support=0.6,
    use_colnames=True
)

print(result)
```

```python
What This Code Does
1 means the item was bought
0 means it was not bought
min_support=0.6 means:
the item must appear in at least 60% of transactions

The program automatically finds popular item combinations.
```
